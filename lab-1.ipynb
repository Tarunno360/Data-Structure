```python
import numpy as np
def meancalcu(arr):
  sum=0
  mean=0
  for i in arr:
    sum+=i
  mean=sum/len(arr)
  return mean
def sdcalcu(arr):
  count1=0
  sd=0
  N1=len(arr)-1
  mean=meancalcu(arr)
  for i in arr:
    count1=count1+(i-mean)**2
  sd=np.sqrt(count1/N1)
  return sd
def newarr(arr):
  diff_count= sdcalcu(arr)*1.5
  count1=0
  for i in arr:
    if abs(i-(meancalcu(arr)))>=diff_count:
      count1+=1
  new=[0]*count1
  count2=0
  for i in range(len(arr)):
    if abs(arr[i]-meancalcu(arr))>=diff_count:
      new[count2]=arr[i]
      count2+=1
  return new
print(format(meancalcu([10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4]),".2f"))
print(format(sdcalcu([10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4]),".2f"))
print(newarr([10, 8, 13, 9, 14, 25, -5, 20, 7, 7, 4]))
```
