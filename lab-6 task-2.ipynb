```python
vou=['a','e','i','o','u','A','E','I','O','U']
def print_alphabets(string,index):
  if index>=len(string):
    return
  st1=string[index]
  if st1 in vou:
    print(f"{index}:{st1}")
  print_alphabets(string,index+1)
  if st1 not in vou:
    print(f"{index}:{st1}")
print_alphabets('Beautiful',0)
```
