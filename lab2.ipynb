```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))

```

Task 1: Zigzag Walk

```python
def walk_zigzag(floor):
  row,column=floor.shape
  for i in range(row):
    if i%2==0:
      for j in range(column):
        if 'W'in floor[i][j]:
          print(floor[i][j],end=' ')
      print()
    else:
      for j in range(column-1,-1,-1):
        if 'W' in floor[i][j]:
          print(floor[i][j],end=' ')
      print()
floor = np.array([[ '3W' , '8B' , '4W' , '6B' , '1W' , '5B'],['3B' , '2W' , '1B' , '6W' , '3B' , '8W'],
                  ['9W' , '0B' , '7W' , '5B' , '3W' , '8B'],['2B' , '1W' , '3B' , '6W' , '0B' , '4W'],
                   ['1W' , '4B' , '2W' , '8B' , '6W' , '6B']])
print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3W 4W 1W
# 8W 6W 2W
# 9W 7W 3W
# 4W 6W 1W
# 1W 2W 6W
```

Task 2: Landscape Screen

```python
def landscape(m):
  row,column=m.shape
  new=np.zeros((column,row),dtype=int)
  for i in range(row):
    for j in range(column):
      new[j][i]=m[i][j]
  return new
m = np.random.randint(15,size=(4,3)) #This creates an np array of 4 x 3 where the values are random numbers from [0,15) range
print_matrix(m)
print()
print_matrix(landscape(m))
```

Task 3: Seating Arrangement

```python
def arrangement(m, diff):
  row,column=m.shape
  answer=True
  for i in range(row):
    for j in range(column):
      if m[i][j]>m[j][i]:
        if m[i][j]-m[j][i]!=diff:
          return False
      elif m[i][j]<m[j][i]:
        if m[j][i]-m[i][j]!=diff:
          return False
      else:
        answer=True
  return answer
m = np.array([[7,13,9,14],[12,8,15,11],[10,17,3,2],[15,10,1,4]])
print_matrix(m)
print()
returned_value = arrangement(m,1)
print(returned_value)#This should print False
unittest.output_test(returned_value, False)
```

Task 4: Chess Piece

```python
def show_knight_move(knight):
  chess_board=np.zeros((8,8),dtype=int)
  pos_move= [(2, 1), (2, -1), (-2, 1), (-2, -1),(1, 2), (1, -2), (-1, 2), (-1, -2)]
  row,column=knight
  for i in pos_move:
    new_row= row+ i[0]
    new_column=column+i[1]
    if 0<=new_row<8 and 0<=new_column<8:
      chess_board[new_row][new_column]=3
  chess_board[knight[0],knight[1]]=33
  return chess_board
knight = (3,4)
chess_board = show_knight_move(knight)
print_matrix(chess_board)
```
