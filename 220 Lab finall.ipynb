# Solver code

```python
# You can possibly write helper functions here
def findPair(arr):
    dic = {}
    idx = 0
    for i in arr:
        if i - 50 in dic:
            return (dic[i - 50], idx)
        if i + 50 in dic:
            return (dic[i + 50], idx)
        dic[i] = idx
        idx += 1
    return (0,0)
```

```python
# Variables: itr, jtr, tSize, tFlag, tScore, sScore, rValue, inpArr
import sys, time, signal, random
sys.setrecursionlimit(9999999)
tScore = 6
sScore = 0
rValue = (0, 0)
tSize = [10, 200, 40000, 60000, 80000, 100000]
def handle_timeout(signum, frame):
    raise TimeoutError
for itr in range(tScore):
    signal.signal(signal.SIGALRM, handle_timeout)
    signal.alarm(2)
    tFlag = True
    inpArr = [0.0 for jtr in range(tSize[itr])]
    for jtr in range(tSize[itr]):
        inpArr[jtr] = random.uniform(1, 99)
    if inpArr[0] < 50:
        inpArr[-1] = inpArr[0] + 50
    else:
        inpArr[-1] = inpArr[0] - 50
    random.shuffle(inpArr)
    try:
        rValue = findPair(inpArr)
    except TimeoutError:
        print('Your function took more than 2 seconds to run on the test number', itr + 1)
        tFlag = False
    finally:
        signal.alarm(0)
    if tFlag == True:
        if inpArr[rValue[0]] - inpArr[rValue[1]] != 50 and inpArr[rValue[1]] - inpArr[rValue[0]] != 50:
            print('Your function returned wrong answer on the test number', itr + 1)
        else:
            print('Your function passed the test number', itr + 1)
            sScore += 1
print('Your score is', sScore, 'out of', tScore)
```
