```python
class Node:
  def __init__(self,elem,next = None):
    self.elem,self.next = elem,next
def createList(arr):
  head = Node(arr[0])
  tail = head
  for i in range(1,len(arr)):
    newNode = Node(arr[i])
    tail.next = newNode
    tail = newNode
  return head

def printLinkedList(head):
  temp = head
  while temp != None:
    if temp.next != None:
      print(temp.elem, end = '-->')
    else:
      print(temp.elem)
    temp = temp.next
  print()
```

```python
def rotates(head):
  temp=head
  while temp.next.next!=None:
    temp=temp.next
  temp2=temp.next
  temp.next=None
  temp2.next=head
  return temp2
head = createList(np.array([10,15,34,41,56,72]))
head = rotates(head)
printLinkedList(head)
```

```python
def common(head1,head2):
  temp1=head1
  temp2=head2
  temp3=None
  while temp1!=None:
    temp2=head2
    while temp2!=None:
      if temp1.elem==temp2.elem:
        if temp3==None:
          temp3=Node(temp1.elem,None)
          temp3_head=temp3
        else:
          temp3.next=Node(temp1.elem)
          temp3=temp3.next
      temp2=temp2.next
    temp1=temp1.next
  return temp3_head
head1 = createList(np.array([1,2,3,4,6]))
head2= createList(np.array([2,4,6,8]))
n=common(head1,head2)
printLinkedList(n)
```

```python
def intersection(l1,l2):
    temp1 = l1

    temp3 = None
    while temp1 != None :

        temp2 = l2
        while temp2 != None:

            if temp1.elem == temp2.elem :
                if temp3==None:

                    temp3 = Node(temp1.elem)
                    new_head = temp3
                else:
                    new_head.next = Node(temp1.elem)
                    new_head = new_head.next

            temp2 = temp2.next
        temp1 = temp1.next

    return temp3
```

```python
first = createList([1, 2, 3, 4, 6])
second = createList([2, 4, 6, 8])
printLinkedList(first)
printLinkedList(second)
x = intersection(second,first)
printLinkedList(x)
```

```python
def leftshift(arr):
  for i in range(1,len(arr)):
    arr[i-1]=arr[i]
  arr[len(arr)-1]=0
  return arr
print(leftshift(np.array([1,2,3,4,5],dtype=int)))
```

```python
def rightshift(arr):
  temp=arr[len(arr)-1]
  for i in range(len(arr)-1,0,-1):
    arr[i]=arr[i-1]
  arr[0]=temp
  return arr
print(rightshift([1,2,3,4,5]))
```
